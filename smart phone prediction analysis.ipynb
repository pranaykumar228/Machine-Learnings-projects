Distribuation and Outliers

```python
a=1
plt.figure(figsize=(15,30))
for i in df[["price","processor_brand","battery_capacity","ram_capacity","internal_memory","screen_size","processor_speed","rating","num_cores"]]:
    if df[i].dtype!="object":
        plt.subplot(9,2,a)
        sns.distplot(df[i])
        plt.subplot(9,2,a+1)
        sns.boxplot(df[i])
        a+=2
```

```python
def remove_outliers(col):
    sorted(col)
    q1,q3=np.quantile(col,[0.25,0.75])
    iqr=q3-q1
    lower_range=q1-(1.5*iqr)
    upper_range=q3+(1.5*iqr)
    return lower_range,upper_range
```

```python
for i in df[["price","processor_brand","battery_capacity","ram_capacity","internal_memory","screen_size","processor_speed","rating","num_cores"]]:
    if df[i].dtype!="object":
        lower,upper=remove_outliers(df[i])
        df[i]=np.where(df[i] > upper,upper,df[i])
        df[i]=np.where(df[i] < lower,lower,df[i])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df["processor_brand"]=le.fit_transform(df["processor_brand"])
df["os"]=le.fit_transform(df["os"])
df["brand_name"]=le.fit_transform(df["brand_name"])
df["has_5g"]=le.fit_transform(df["has_5g"])
df["has_nfc"]=le.fit_transform(df["has_nfc"])
df["model"]=le.fit_transform(df["model"])
df["has_ir_blaster"]=le.fit_transform(df["has_ir_blaster"])
```
